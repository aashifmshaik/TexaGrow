```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
days = 30  # Duration of the experiment in days
moisture_levels = np.arange(10, 101, 10)  # Moisture levels from 10% to 100%
optimal_moisture = 60  # Optimal moisture level for plant growth (in %)
growth_rate_constant = 0.05  # Growth rate multiplier for optimal conditions

# Growth function: simulates growth rate based on moisture
def calculate_growth(moisture, day):
    deviation = abs(moisture - optimal_moisture)
    growth_rate = growth_rate_constant * np.exp(-0.1 * deviation)  # Adjusted for moisture deviation
    return growth_rate * day  # Growth per day increases with time if moisture is close to optimal

# Run simulation and store results
growth_data = {}  # To store plant growth at each moisture level

for moisture in moisture_levels:
    daily_growth = []
    for day in range(1, days + 1):
        daily_growth.append(calculate_growth(moisture, day))
    growth_data[moisture] = daily_growth

# Plotting the results
plt.figure(figsize=(10, 6))
for moisture, growth in growth_data.items():
    plt.plot(range(1, days + 1), growth, label=f'{moisture}% Moisture')

plt.xlabel('Days')
plt.ylabel('Plant Growth')
plt.title('Plant Growth Simulation at Different Soil Moisture Levels')
plt.legend(title='Moisture Level')
plt.show()
```
